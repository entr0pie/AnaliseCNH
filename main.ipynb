### Criando a coluna target para a base inteira

- Primeiro, vamos importar a base:

```python
import pandas as pd

df = pd.read_excel('./assets/Lista NPS Positivo_V4.xlsx', sheet_name='answers')
df.head()
```

- Depois, vamos criar uma nova coluna `target`, com as seguintes categorias:
    - Promotor: nota entre 9 e 10;
    - Neutro: nota entre 7 e 8;
    - Detrator: nota menor que 7; 

```python
# https://stackoverflow.com/a/26887820

def create_nps_target(row: pd.Series) -> str:
    nota = row['nota']
    
    if nota >= 9:
        return 'Promotor'
    if nota >= 7:
        return 'Neutro'

    return 'Detrator'

df['target'] = df.apply(create_nps_target, axis=1)
df.head()
```

- Por fim, podemos contabilizar cada uma das classes:

```python
value_counts_target = df['target'].value_counts()
value_counts_target

assert value_counts_target['Promotor'] == 18251
assert value_counts_target['Neutro'] == 4738
assert value_counts_target['Detrator'] == 2185
```

### Volumetria do Grupo 4: Promotores, Neutros e Detratores

```python
grupo_4_value_counts = df['target'].value_counts()
grupo_4_total_count = grupo_4_value_counts.sum()

percentage = (grupo_4_value_counts / grupo_4_total_count) * 100

pd.DataFrame({
    'target': ['Promotor', 'Neutro', 'Detrator'],
    'quantidade': [grupo_4_value_counts['Promotor'], grupo_4_value_counts['Neutro'], grupo_4_value_counts['Detrator']],
    'porcentagem (%)': [percentage['Promotor'], percentage['Neutro'], percentage['Detrator']]
})
```

### Criando a coluna de região

```python
# https://stackoverflow.com/a/26887820

def create_regiao_column(row: pd.Series) -> str | None:
    regioes = {
        'Sul': ['RS', 'SC', 'PR'],
        'Sudeste': ['SP', 'RJ', 'MG', 'ES'],
        'Centro-Oeste': ['MT', 'MS', 'GO', 'DF'],
        'Nordeste': ['SE', 'RN', 'PI', 'PE', 'PB', 'MA', 'CE', 'BA', 'AL'],
        'Norte': ['TO', 'RR', 'RO', 'PA', 'AM', 'AP', 'AC']
    }

    for regiao, siglas in regioes.items():
        if row['estado'] in siglas:
            return regiao
    
    return None

df['regiao'] = df.apply(create_regiao_column, axis=1)
df.head()
```
